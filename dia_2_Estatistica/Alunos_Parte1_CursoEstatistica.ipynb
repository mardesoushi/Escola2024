```python
import math as mt

def binomial(n, p):
    return mt.factorial(n) / (mt.factorial(n - p) * mt.factorial(p))

probe = 0.4
k_sucess = 3

## Probabilidade de ocorrerem 3 sucessos
def negative_binominal(probe, k_sucess, n_tries):
    prob= binomial(n_tries - 1, k_sucess - 1) * probe ** k_sucess * (1 - probe) ** (n_tries - k_sucess)
    return prob

prob_list = []
for n_tries in range(k_sucess+1, 50):
    prob_list.append(negative_binominal(probe, k_sucess, n_tries))

plt.plot(range(k_sucess+1, 50), prob_list) #prob_list
```

## **Distribuição Binomial**

Plote a curva de uma distribuição Binomial para diferentes valores de probabilidade

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Definindo os parâmetros
N = 20  # Número de ensaios
p_values = np.arange(0.1, 1.0, 0.2)  # Variando p de 0 a 1 em incrementos de 0.1
k = np.arange(0, N + 1)  # Possíveis números de sucessos

# Criando a figura
plt.figure(figsize=(12, 8))

# Loop para cada valor de p
for p in p_values:
    # Calculando as probabilidades binomiais
    probabilities = binom.pmf(k, N, p)

plt.plot(k, probabilities)

```

# **Distribuição Gaussiana**

Faça um gráfico da distribuição gaussiana

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Definindo os parâmetros da Gaussiana
mu = 0       # média
sigma = 1    # desvio padrão

# Gerando valores x para a função Gaussiana
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y)
```
